### ¿Qué es el método de las diferencias divididas?

Es un método útil para generar sucesivamente los polinomios dado un conjunto de puntos de una función supuestamente derivable.

#### Características:
+ Nos dara como resultado un polinomio de grado n-1 puntos que queremos aproximar.

El presente cuadernillo presenta un ejemplo para aproximar una seríe de puntos aunque puede ser modificado para acoplarlo a tus necesidades.

```python
import sympy as sym #Usaremos SymPy para computo simbólico.

import numpy as np #Usaremos numpy para arreglos de números.

import pandas as pd #Usaremos Pandas para crear y manejar data frames donde registraremos los datos.

import IPython.display as disp #Usaremos el display para visualizar ecuaciones.
```

```python
# Ingresamos los datos de los cuales queremos sacar el polinomio.

xi = np.array([-2, -1, 0, 2, 3, 6])
fi = np.array([-18, -5, -2, -2, 7, 142])


# Empezamos a crear nuestra tabla de diferencias dividas, solo le damos forma a nuestra matriz

titulo = ['i   ','xi  ','fi  ']
n = len(xi)
ki = np.arange(0, n, 1)
tabla = np.concatenate(([ki],[xi],[fi]),axis=0)

tabla = np.transpose(tabla)

# Aqui iniciamos la parte de la matriz en 0 de tipo flotante para empezar a calcularla

dfinita = np.zeros(shape=(n,n),dtype=float)
tabla = np.concatenate((tabla,dfinita), axis=1)


# Guardamos en esta lista la forma que tendra nuestra tabla de diferencias divididas
[n,m] = np.shape(tabla)

# Declaramos donde iniciara nuestra tabla
diagonal = n-1
j = 3
while (j < m):
    # Añade título para cada columna de la tabla para saber que diferencia dividida es
    titulo.append('F['+str(j-2)+']')

    # cada fila de columna
    i = 0
    paso = j-2 # inicia en 1
    #Aqui empezamos a calcular nuestras diferencias divididas
    
    while (i < diagonal):
        denominador = (xi[i+paso]-xi[i])
        numerador = tabla[i+1,j-1]-tabla[i,j-1]
        tabla[i,j] = numerador/denominador
        i = i+1
    diagonal = diagonal - 1
    j = j+1

    
# Aquí empezamos a construir nuestra tabla de diferencias divididas:

dDividida = tabla[0,3:]
n = len(dfinita)


x = sym.Symbol('x') #Declaramos nuestro simbolo para usar con sympy
polinomio = fi[0]

for j in range(1,n,1):
    factor = dDividida[j-1]
    termino = 1
    for k in range(0,j,1):
        termino = termino*(x-xi[k])
    polinomio = polinomio + termino*factor

# simplifica multiplicando entre (x-xi)

polisimple = polinomio.expand()
np.set_printoptions(precision = 4)

print('Tabla Diferencia Dividida \n')
print([titulo])
print(tabla)
print('\n')

print('dDividida: ')
print(dDividida)
print('\n')

print('polinomio: ')
print(polinomio)
print('\n')

print('polinomio simplificado: ' )
print(polisimple)
```
